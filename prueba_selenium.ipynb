
La solución podría ser asignar tipo de variable para cada fila y, si no es el dato de ese tipo, pasar al campo siguiente. En este caso, las tres primeras son de tipo string, pero la última es siempre de tipo numérico

```python
buscar = "Vicente+Sisto"
#iniciamos el driver

driver = webdriver.Chrome(driver_path, chrome_options = options)

#aquí declaramos la web a la que entraremos

driver.get('https://scholar.google.com/citations?view_op=search_authors&mauthors={}&hl=es&oi=drw'.format(buscar))

#iniciamos el proceso de búsqueda

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'span.gs_hlt')))\
    .click()

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'span#gsc_a_ha.gsc_a_h')))\
    .click()

WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                     'tbody#gsc_a_b')))\

publicaciones = driver.find_element(by=By.CSS_SELECTOR, value='tbody#gsc_a_b')
publicaciones = publicaciones.text

#en este caso, debemos hacer el split con "\n" porque es un caracter no imprimible de LineFeed
#aquí presentan cómo se usa https://www.geeksforgeeks.org/python-string-splitlines-method/
#esto permite visualizar los caracteres unicode no imprimibles https://www.soscisurvey.de/tools/view-chars.php

listado = publicaciones.split("\n")

titulos = list()
autores = list()
referencia = list()
año = list()

for i in range(0, len(listado), 4):
    if type(listado[i]) == str:
        titulos.append(listado[i])
    else:
        titulos.append("NaN")
    autores.append(listado[i+1])
    referencia.append(listado[i+2])
    año.append(listado[i+3])

df = pd.DataFrame({"titulos" : titulos, "autores" : autores, "referencia" : referencia, "año" : año})

#print(publicaciones)
#print(listado)
df
```
