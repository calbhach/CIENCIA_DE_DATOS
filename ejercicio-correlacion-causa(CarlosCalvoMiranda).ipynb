```python
sns.pairplot(df, hue="Brand")
```

```python
#Extraeremos dos variables que tienen poca relaciÃ³n teÃ³rica, bmi y cantidad de hijos
df_espurio = pd.DataFrame([df["Listing Price"], df["Sale Price"], df["Discount"], df["Rating"], df["Reviews"] ]).transpose()
df_espurio
```

```python
scaler_p = StandardScaler()
scaled_p = scaler_p.fit_transform(df_espurio)
scaled_p.T.shape
covariance_matrix_p = np.cov(scaled_p.T) # Datos Traspuestos
covariance_matrix_p
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm_p = sns.heatmap(covariance_matrix_p,
                cbar=True,
                annot=True,
                square=True,
                annot_kws={'size':12},
                yticklabels=["Listing Price", "Sale Price", "Discount", "Rating", "Reviews"],
                xticklabels=["Listing Price", "Sale Price", "Discount", "Rating", "Reviews"]
                )
```

Podemos notar que existe la misma potencia de correlaciÃ³n entre las variables precio de venta y precio de lista, y de cantidad de descuento y cantidad de reviews. 
Si quisiÃ©semos confundir correlaciÃ³n con causa, podemos decir que una mayor cantidad de revisiones sobre el producto genera mayores descuentos, del mismo modo que los productos mÃ¡s caros terminan vendiÃ©ndose mÃ¡s caros pese al descuento.
